```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
```

```python
top_stocks = {'Stocks':['AAPL', 'MSFT', 'AMZN', 'GOOG', 'TSLA', 'FB', 'NVDA', 'PYPL', 'NFLX', 'CMCSA', 'INTC', 'ADBE', 'PEP', 'CSCO',
              'AVGO', 'QCOM', 'TMUS', 'COST', 'TXN', 'AMGN']}
```

```python
#Extracting data from yahoo finance
```

```python
df={}
for i in range(0,20):
    df[top_stocks['Stocks'][i]] = yf.Ticker(top_stocks['Stocks'][i]).history(top_stocks['Stocks'][i] , start = "2010-01-01", end = "2020-01-01")['Close']
```

```python
data_returns=pd.DataFrame(df).pct_change()
data_returns
```

```python
#Finding Correlation Matrix
```

```python
corr_matrix = data_returns.corr(method='pearson')
corr_matrix
```

```python
#Sorting and find 5 most uncorrelated stocks
```

```python
df1 = pd.DataFrame(corr_matrix.min()) 
df2 = df1.sort_values(by=[0], axis=0, ascending=True, inplace=False, kind='quicksort', na_position='last', ignore_index=False, key=None)
df2
```

```python
print("5 most uncorrelated stocks are ",df2.index[0],df2.index[1],df2.index[2],df2.index[3],df.index[4])
```
